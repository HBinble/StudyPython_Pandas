```python
# split()과 concat()으로 데이터 정리

ebola = pd.read_csv('./data/country_timeseries.csv')
print(ebola.columns)
```

```python
print(ebola.iloc[:5, [0,1,2,3,10,11]])
```

```python
# Date, Day 열 고정 / 나머지 열 피벗
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

```python
# 'variable' 열의 데이터를 _ 기준으로 분리
variable_spilt = ebola_long.variable.str.split('_') #ebola_long['variable'].str.split('_')
variable_spilt[:5]
```

```python
status_values = variable_spilt.str.get(0)
country_values = variable_spilt.str.get(1)

print(status_values[:5])
print(status_values[-5:])
# status_values.head(), status_values.tail()

print(country_values[:5])
print(country_values[-5:])
#country_values.head(), country_values.tail()
```

```python
# 분리된 데이터를 Data.frame에 저장
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long.head()
```

```python
# 분리된 데이터를 concat()으로 합치기
variable_spilt = ebola_long.variable.str.split('_', expand=True)
variable_spilt.columns = ['status2', 'country2']
ebola_parsed = pd.concat([ebola_long, variable_spilt], axis=1)
ebola_parsed
```

```python
ebola = ebola_parsed.loc[:, ['Date', 'Day', 'status', 'country', 'value']]
ebola
```

```python
ebola.dropna()
```

```python
weather = pd.read_csv('./data/weather.csv')
weather.head()
```

```python
print(weather.shape)
print(weather.columns)
print(weather.iloc[:5, :11])
```

```python
weather_melt = pd.melt(weather,
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day', value_name='temp')
weather_melt[100:110]
```

```python
# pivot_table() 사용(행 데이터를 열 데이터로)

weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], # element 열의 데이터 tmax/tmin 를 열로 변환 
    columns='element', values='temp') # tmax/ tmin 열의 데이터는 temp

weather_tidy.head()
```

```python
# 인덱스 초기화
weather_tidy_falt = weather_tidy.reset_index()
weather_tidy_falt.head()
```

```python
# 빌보드 차트 중복 데이터 처리하기(정규화)

billboard = pd.read_csv('./data/billboard.csv')
billboard
```

```python
billboard_long = pd.melt(
    billboard,
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
    var_name='week', value_name='rating')

print(billboard_long.shape)
billboard_long.head(10)
```

```python
billboard_long[billboard_long.track == 'Loser'].head()
```

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time', 'date.entered']]
print(billboard_songs.shape)
```

```python
# 중복 데이터 제거(drop_duplicates())
billboard_songs = billboard_songs.drop_duplicates() # billboard_songs.drop_duplicates(inplace=True)
print(billboard_songs.shape)
```

```python
# 순번(id) 추가
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head(10)
```

```python
billboard_ratings = billboard_long.merge(billboard_songs, 
                                         on=['year', 'artist', 'track', 'time'])
print(billboard_ratings.shape)
billboard_ratings.head()
```

```python
billboard_ratings = billboard_songs.merge(billboard_long, 
                                         on=['year', 'artist','track','time','date.entered'])
```
