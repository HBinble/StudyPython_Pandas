### 누락값 확인하기


```python
import numpy as np
```

```python
print(np.NaN == True)
print(np.NaN == False)
print(np.NaN == 0)
print(np.NaN == '')
```

```python
print(np.NaN == np.NaN)
print(np.NaN == np.nan)
print(np.NaN == np.NAN)
print(np.nan == np.NAN)
```

```python
import pandas as pd
```

```python
# 누락값을 확인하는 isnull / notnull 메소드

print(pd.isnull(np.NaN))
print(pd.isnull(np.nan))
print(pd.isnull(np.NAN))

print(pd.notnull(np.NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

##### 누락값 개수 확인 : count_nonzero(), isnull(), value_counts()

```python
import numpy as np
```

```python
# count_nonzero() : 배열에서 0(False)이 아닌 값의 개수 확인

print(np.count_nonzero(ebloa.isnull()))
print(np.count_nonzero(ebloa['Cases_Guinea'].isnull()))
```

```python
# value_counts() : 지정한 열의 빈도 확인

print(ebloa['Cases_Guinea'].value_counts(dropna=False))
```

```python
ebloa['Cases_Guinea'].value_counts()
```
